```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import KernelPCA
```

```python
def generate_data_sphere(radius, num_samples, noise):
  theta = np.random.uniform(0, np.pi, num_samples)
  phi = np.random.uniform(0, 2*np.pi, num_samples)

  x = radius * np.sin(phi) * np.cos(theta)
  y = radius * np.sin(phi) * np.sin(theta)
  z = radius * np.cos(phi)

  x += np.random.normal(0, noise, num_samples)
  y += np.random.normal(0, noise, num_samples)
  z += np.random.normal(0, noise, num_samples)

  return np.array([x, y, z]).T
```

```python
num_samples = 1000
inner_radius = 0.5
outer_radius = 1
sigma_noise = 0.05

inner_sphere = generate_data_sphere(radius=inner_radius, num_samples=num_samples, noise=sigma_noise)
outer_sphere = generate_data_sphere(radius=outer_radius, num_samples=num_samples, noise=sigma_noise)

x = np.vstack([inner_sphere, outer_sphere])

rbf_kernel = KernelPCA(kernel='rbf', gamma=5, degree=7, n_components=2)
x_pca = rbf_kernel.fit_transform(x)

plt.scatter(x_pca[:, 0], x_pca[:, 1], c=np.concatenate([np.zeros(num_samples), np.ones(num_samples)]))
plt.title("RBF kernel")
plt.xlabel("C1")
plt.ylabel("C2")
plt.show()
```

```python
rbf_kernel = KernelPCA(kernel='poly', gamma=5, degree=7, n_components=2)
x_pca = rbf_kernel.fit_transform(x)

plt.scatter(x_pca[:, 0], x_pca[:, 1], c=np.concatenate([np.zeros(num_samples), np.ones(num_samples)]))
plt.title("7 degree Kernel")
plt.xlabel("C1")
plt.ylabel("C2")
plt.show()
```

```python
from sklearn.datasets import fetch_lfw_people
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
```

```python
lfw_people = fetch_lfw_people(min_faces_per_person=50)
X =lfw_people.data
y = lfw_people.target
names = lfw_people.target_names
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.fit_transform(X_test)
```

```python
n_components = 100
pca_rbf = KernelPCA(kernel='rbf')
X_train_pca_100 = pca_rbf.fit_transform(X_train_scaler)[:,:n_components]
X_test_pca_100 = pca_rbf.transform(X_test_scaler)[:,:n_components]

n_components = 50
X_train_pca_50 = pca_rbf.fit_transform(X_train_scaler)[:,:n_components]
X_test_pca_50 = pca_rbf.transform(X_test_scaler)[:,:n_components]
```

```python
mlpc = MLPClassifier(random_state=1, max_iter=300)
mlpc.fit(X_train_pca_100,y_train)
mlpc_pred = mlpc.predict(X_test_pca_100)

# MLP for 100 components
print(classification_report(y_test, mlpc_pred, target_names=names))
```

```python
mlpc = MLPClassifier(random_state=1, max_iter=300)
mlpc.fit(X_train_pca_50,y_train)
mlpc_pred = mlpc.predict(X_test_pca_50)

# MLP for 50 components
print(classification_report(y_test, mlpc_pred, target_names=names))
```

```python
svc = SVC(random_state=42)
svc.fit(X_train_pca_100, y_train)
svc_pred = svc.predict(X_test_pca_100)

# SVC for 100 components
print(classification_report(y_test, svc_pred, target_names=names))
```

```python
svc = SVC(random_state=42)
svc.fit(X_train_pca_50, y_train)
svc_pred = svc.predict(X_test_pca_50)

# SVC for 50 components 
print(classification_report(y_test, svc_pred, target_names=names))
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_pca_100, y_train)
rf_predictions = rf.predict(X_test_pca_100)

# Random forest for 100 components
print(classification_report(y_test, rf_predictions, target_names=names))
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_pca_50, y_train)
rf_predictions = rf.predict(X_test_pca_50)

# Random forest for 50 components
print(classification_report(y_test, rf_predictions, target_names=names))
```

```python
n_components = 100
pca_poly_100 = KernelPCA(kernel='poly')
X_train_pca_100 = pca_poly_100.fit_transform(X_train_scaler)[:,:n_components]
X_test_pca_100 = pca_poly_100.transform(X_test_scaler)[:,:n_components]

n_components = 50
pca_poly_50 = KernelPCA(kernel='poly')
X_train_pca_50 = pca_poly_50.fit_transform(X_train_scaler)[:,:n_components]
X_test_pca_50 = pca_poly_50.transform(X_test_scaler)[:,:n_components]
```
