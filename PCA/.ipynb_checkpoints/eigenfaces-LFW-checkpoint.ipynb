```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import fetch_lfw_people
```

```python
lfw_people = fetch_lfw_people(min_faces_per_person=50)
X =lfw_people.data
y = lfw_people.target
names = lfw_people.target_names
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.fit_transform(X_test)
```

```python
pca = PCA()
pca.fit(X_train)
n_components = 100
X_train_pca = pca.transform(X_train_scaler)[:, :n_components]
X_test_pca = pca.transform(X_test_scaler)[:, :n_components]
```

```python
from sklearn.svm import SVC, SVR
```

```python
svm = SVC(kernel='linear', C=1)
svm.fit(X_train_pca, y_train)
```

```python
svm_predictions = svm.predict(X_test_pca)
```

```python
from sklearn.metrics import classification_report
```

```python
# PCA for 100 components
print(classification_report(svm_predictions, y_test, target_names=names))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_pca, y_train)
rf_predictions = rf.predict(X_test_pca)


# Random forest for 100 components
print(classification_report(y_test, rf_predictions, target_names=names))
```

```python
from sklearn.neural_network import MLPClassifier

mlpc = MLPClassifier(random_state=1, max_iter=300)
mlpc.fit(X_train_pca,y_train)
mlpc_pred = mlpc.predict(X_test_pca)

# MLP for 100 components
print(classification_report(y_test, rf_predictions, target_names=names))
```

```python
eigenfaces = pca.components_[:20]
fig, axes = plt.subplots(4, 5, figsize=(10, 10))
axes = axes.ravel()

for i, ax in enumerate(axes):
    ax.imshow(eigenfaces[i].reshape(62, 47), cmap='gray')
    ax.axis('off')

plt.show()

# Eigenfaces stanowią niskowymiarową reprezentację zbioru danych, umożliwiając na ich podstawie zredukować elementy zbioru
# Eigenfaces reeprezentują cechy związane z padającym oświetleniem i podstawowe cechy twarzy takie jak kształt nosa, rozstaw oczu czy usta 
```
